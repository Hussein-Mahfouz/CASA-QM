```python
import numpy as np
import pandas as pd
```

```python
# read in results. header = None used so that first row is not treated as a header
df_results = pd.read_csv("allocation.csv", header = None)
df_results.head()
```

## Get the names of the garages in a list to append as a header to the results

```python
garages_df = pd.read_csv("distance_matrix.csv")
# remove 1st column which is the terminal associated with the garage
del garages_df['InputID']
garages_df.head()
```

## Extract header as list

```python
garages = list(garages_df.columns.values)
garages
```

```python
## Add list as headers using .columns
```

```python
df_results.columns = garages
df_results.head()
```

```python
df_results.sum(axis = 0) 
```

```python
df_summary = df_results.sum(axis = 0) 
type(df_summary)
```

```python
### we need to convert it to a dataframe
```

```python
df_summary = df_summary.to_frame()
```

### Reset index and add column names

```python
# reset index so that Terminal Name is a Column, not the index
df_summary.reset_index(inplace=True)
#Add Column headers (to merge with later)
#headers = ['Name', 'Buses Allocates']
df_summary.columns = ['Name', 'Buses_Allocated']
df_summary
```

### Add the Garage Capacities csv to merge them together

#### .str.strip() removes whitespace. Otherwise we are unable to merge

```python
df_capacities = pd.read_csv("Garage Capacities.csv")
# strip function - removes all whitespace from all stringlike cells in dataframe
# https://stackoverflow.com/questions/33788913/pythonic-efficient-way-to-strip-whitespace-from-every-pandas-data-frame-cell-tha
df_capacities = df_capacities.apply(lambda x: x.str.strip() if x.dtype == "object" else x)
```

### Merge

```python
# how = 'left' joins over the left dataframe, in this case we are joining onto the df_trips dataframe
df_plot = pd.merge(df_summary, df_capacities, left_on='Name', right_on='Name', how='left')
df_plot
```

## MORE PREP FOR R PLOTS

#### Average Dead Mileage Per Trip

```python
# read in results. header = None used so that first row is not treated as a header
df_deadMile = pd.read_csv("mileage_per_trip.csv", header = None)
df_deadMile.head()
```

#### Convert to Numpy array to get the maximum value from each row (the only non-zero value)

```python
np_deadMile = df_deadMile.values
np_deadMile
```

#### Get dead mileage for each trip

```python
deadMileAll = np.amax(np_deadMile, axis=1)   # Maxima along the second axis (rows)
```

```python
df_deadMilePlot = pd.DataFrame(deadMileAll)
df_deadMilePlot 
```

# Get number of buses on each trip - to append to the dataframe above. This way we can plot a histogram for the dead mileage of all buses, not just all trips

```python
# read in the csv
df_buses = pd.read_csv("bus-to-garage.csv")
# keep only the buses column
df_buses = df_buses[["standard_buses"]]
df_buses
```

# Merge by index

```python
merged = pd.merge(df_deadMilePlot,df_buses,left_index=True,right_index=True)
merged
```
