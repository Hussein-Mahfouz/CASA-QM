```python
import pandas as pd
import numpy as np
```

```python
# import the distance matrix csv as a dataframe 
df = pd.read_csv("terminal-garage-distance-v2.csv")

df.head()
```

### Multiply Distance by 2 to account for Pull in and Pull out

```python
df.Distance = df.Distance * 2
df
```

```python
# Pivot DataFrame so that the garages are columns
# https://stackoverflow.com/questions/44513488/how-to-remove-multilevel-index-in-pandas-pivot-table
df = df.pivot(index='InputID', columns='TargetID', values = 'Distance').reset_index().rename_axis(None, axis=1)
df.head()
```

```python
# keep only the garages used by the CTA buses. There are four garages that are used by CTA minibuses (these are dropped)
df.drop(['Katameya', 'Al Mataria', 'El Basatin', 'Tiba'], axis = 1, inplace = True)
df.head()
```

```python
print(df.loc[0, 'InputID'])
```

#### save file to csv

```python
df.to_csv(r'distance_matrix.csv', index = None, header=True)
```
